**Pandas**

Es una poderosa herramienta utilizada para el análisis y manipulación de datos. Ofrece estructuras de datos flexibles y eficientes, como DataFrames, que permiten trabajar con conjuntos de datos tabulares de manera fácil y eficiente. Algunas de las principales funcionalidades de Pandas incluyen:



1.   **DataFrames**: Estructuras bidimensionales que permiten almacenar y manipular datos en forma de tablas con filas y columnas.
2.   **Manipulación de datos**: Ofrece funciones para filtrar, seleccionar, agrupar y transformar datos, facilitando tareas como la limpieza y preparación de conjuntos de datos.
3.   **Manejo de datos faltantes:** Pandas proporciona métodos para manejar valores nulos o faltantes, lo que es esencial en el análisis de datos del mundo real.
4.   **Operaciones eficientes:** Permite realizar operaciones vectorizadas y eficientes en grandes conjuntos de datos, evitando la necesidad de bucles for.
3.   **Entrada/salida de datos:** Facilita la lectura y escritura de datos en varios formatos, como CSV, Excel, SQL, entre otros.
4.   **Visualización de datos:** Se integra fácilmente con bibliotecas de visualización como Matplotlib y Seaborn para crear gráficos y visualizaciones informativas.








```python
import pandas as pd
```

**Matplotlib**

Es una herramienta fundamental para la visualización de datos. Su propósito principal es facilitar la creación de gráficos y visualizaciones a partir de conjuntos de datos. Algunas de las funciones clave de Matplotlib incluyen:


1.   **Gráficos estáticos y dinámicos:** Permite la generación de una amplia variedad de gráficos estáticos, como gráficos de líneas, barras, dispersión, histogramas, entre otros. También es compatible con la creación de gráficos interactivos.
2.   **Personalización:**  Brinda un alto grado de control sobre la apariencia de los gráficos, lo que permite ajustar colores, estilos, etiquetas y otros elementos para adaptarse a las necesidades específicas de visualización.
3.   **Subgráficos y diseño:** Facilita la creación de subgráficos y la disposición de múltiples gráficos en una sola figura, permitiendo comparar diferentes conjuntos de datos de manera efectiva.
4.   **Compatibilidad con varios formatos de salida:** Los gráficos pueden guardarse en varios formatos, como PNG, PDF, SVG, lo que facilita la inclusión de visualizaciones en informes, presentaciones o páginas web.
5.   **Integración con Pandas:** Se integra bien con la librería Pandas, lo que facilita la visualización de datos almacenados en estructuras de datos Pandas, como DataFrames.




```python
import matplotlib.pyplot as plt
```

**NumPy**

Es esencialmente utilizada para realizar operaciones numéricas eficientes y trabajar con arreglos multidimensionales. Algunas de las funcionalidades clave de NumPy incluyen:

1.   **Arreglos multidimensionales (ndarrays)**: NumPy proporciona una estructura de datos eficiente para almacenar y manipular grandes conjuntos de datos en forma de arreglos multidimensionales, lo que facilita el manejo de datos numéricos y científicos.
2.   **Operaciones matemáticas:** Ofrece un conjunto extenso de funciones y operaciones matemáticas que pueden aplicarse directamente a los elementos de los arreglos, permitiendo cálculos rápidos y eficientes.
3.   **Broadcasting:** Facilita la realización de operaciones entre arreglos de diferentes formas y tamaños de manera automática, simplificando la escritura de código y mejorando la eficiencia computacional.
4.   **Funciones de álgebra lineal:** Incluye funciones para realizar operaciones fundamentales de álgebra lineal, como multiplicación de matrices, descomposiciones, resolución de sistemas de ecuaciones lineales, entre otras.
5.   **Generación de números aleatorios:** NumPy proporciona herramientas para generar secuencias de números aleatorios, útiles en simulaciones y experimentos estadísticos.
6. **Integración con otros paquetes:** Se integra fácilmente con otras bibliotecas científicas y de análisis de datos en Python, como SciPy, Pandas y Matplotlib, para proporcionar un ecosistema completo para el cómputo científico.

5. **Realizar un "pair plot"**: Pairplot es una herramienta que permite crear gráficos por **pares** de características. Además de manera más interesante, su diagonal permite visualizar el histograma o densidad que permite ver la distribución univaridad para cada característica.

Osbervar que fue necesario importar la librería **seaborn** que es similar a matplotlib, pero me permite generar este pairplot.

Por otro lado, verificar que para poder distinguir cada clase, se uso el atributo "hue" especificando a "Species" como mi variable de clases y de la misma manera, los marcadores:

* "o" circulos
* "s" cuadrados
* "D" diamantes

Finalmente, se escogió una paleta de colores "Set2", puedes investigar y cambiarla si deseas.


```python
import seaborn as sns
sns.pairplot(Iris, hue="Species", markers=["o", "s", "D"], palette="Set2")
```
